```python
from selenium import webdriver
from selenium.common.exceptions import WebDriverException
import chromedriver_binary
import datetime
import time
import pyautogui as pgui
import pyperclip

def now_time():
   '''
   現在の時間を表示する関数
   ログの出力で使う

   引数：
   　なし
   戻り値：
   　現時点の時刻　mm/dd HH:MM
   '''
   dt_now = datetime.datetime.now()
   return dt_now.strftime('%m/%d %H:%M')+' '


class Instagram():
   '''
   インスタグラムの自動投稿をするクラス
   つぎの４つの機能がある
   ・ログイン:login()
   ・ユーザー画面への移動:to_user_page()
   ・画像の投稿:post_pic()
   ・処理を終了:quit()
   '''
   def __init__(self, username, password):
       # super(Instagram, self).__init__()
       #スマホ画面のブラウザ
       mobile_emulation = { "deviceName": "Galaxy S5" }
       options = webdriver.ChromeOptions()
       options.add_experimental_option("mobileEmulation", mobile_emulation)

       self.driver = webdriver.Chrome(options=options)
       self.username = username
       self.password = password   

       print(now_time()+'ブラウザを開きました')

   def login(self):
       '''
       インスタグラムにアクセスして、ログインする関数
       
       引数、戻り値なし
       '''
       self.driver.get('https://www.instagram.com/accounts/login/?source=auth_switcher')
       print(now_time()+'instagramにアクセスしました')
       time.sleep(1)
       
       #メアドと、パスワードを入力
       self.driver.find_element_by_name('username').send_keys(self.username)
       time.sleep(1)
       self.driver.find_element_by_name('password').send_keys(self.password)
       time.sleep(1)

       #ログインボタンを押す
       self.driver.find_elements_by_xpath('//button[@type="submit"]')[0].click()
       time.sleep(3)
       print(now_time()+'instagramにログインしました')

       time.sleep(1)

   def to_user_page(self):
       '''
       ユーザー画面への移動に移動する関数
       
       引数、戻り値なし
       '''
       self.driver.get('https://www.instagram.com/' + self.username)

   def post_pic(self, file_path, caption):
       '''
       画像を投稿する関数
       
       引数：
       ・投稿するファイルのパス
       ・投稿するキャプション
       
       戻り値：
       　なし
       '''
       #アップロードボタンをクリック
       self.driver.find_elements_by_xpath('//div[@class="q02Nz _0TPg"]')[0].click()
       time.sleep(1)

        #画像を選択        pgui.hotkey('command', 'shift', 'g')
       time.sleep(0.5)

       pyperclip.copy(file_path)
       pgui.hotkey('command', 'v')
       time.sleep(0.5)

       pgui.hotkey('return')

       time.sleep(1)
       pgui.hotkey('return')
       time.sleep(2)

       #「次へ」ボタンをクリックする
       self.driver.find_elements_by_xpath('//button[@class="UP43G"]')[0].click()
       time.sleep(2)

       #キャプション入力
       self.driver.find_elements_by_xpath('//textarea[@class="_472V_"]')[0].send_keys(caption)
       time.sleep(3)
       
       #「シェア」ボタンをクリックする
       self.driver.find_elements_by_xpath('//button[@class="UP43G"]')[0].click()
       time.sleep(3)

       print(now_time()+'投稿しました')


       
   def quit(self):
       '''
       処理を終了する関数
       
       引数、戻り値なし
       '''
       
       self.driver.close()
       self.driver.quit()
       print(now_time()+'ブラウザを閉じました')
               
def main():

   username = '{ここにインスタアカウントのユーザー名を入力}'
   password = '{ここにインスタアカウントのパスワードを入力}'
   file_path = '{ここに画像のディレクトリを入力}'
   #例 file_path = '/Users/{ユーザー名}/Desktop/test.jpg'
   caption = '{ここにキャプション（投稿テキスト）を入力}'
   #例 caption = '#テスト \n#スマホ'

   my_bot = Instagram(username, password)

   try:
       my_bot.login()
       time.sleep(2)

       my_bot.to_user_page()
       time.sleep(2)

       my_bot.post_pic(file_path, caption)
       time.sleep(2)
       
   except WebDriverException as e:
       print(e)

   my_bot.quit()

if __name__ == '__main__':
   main()
```
